```python
import pandas as pd
import numpy as np

filename = 'world-happiness-report-2019.csv'
df = pd.read_csv(filename)
df.iloc[:,:-1].values
```

```python
df.info()
```

```python
df = df.drop_duplicates()
df = df.dropna()
```

```python
df.info()
```

```python
#second way
fill = pd.read_csv(filename)
drop  = fill.drop(columns='Country (region)')
df_filled = drop.fillna(drop.mean())
df_filled.info()
```

```python



def remove_outliers(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df_no_outliers = df[((df >= lower_bound) & (df <= upper_bound)).all(axis=1)]
    
    return df_no_outliers


out = remove_outliers(df_filled)

```

```python

statistics = df.describe()
print(statistics)


```

```python

mean = out.mean()
median = out.median()
std = out.std()
min_values = out.min()
max_values = out.max()

print(" mean : ")
print(mean)

print("median :")
print(median)

print("std :")
print(std)

print("min_values :")
print(min_values)

print("max_values :")
print(max_values)


```
