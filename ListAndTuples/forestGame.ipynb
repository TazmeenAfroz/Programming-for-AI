```python
list = ['friendlyCreature','deadlyCreature']
def forest():
    print('Welcome to the lost forest')
    print(' you have to find the friendly creature to get out of the forest')
    print(' But beware of the deadly creature')
    print('you have only 4 directions to go') 
    print('you will find many treasures in this journey and you will get bonus points')
    print('you can use these points to buy the weapon to kill the deadly creature')
    print('you will also find the life points in this journey')
    print('Now lets start the journey...........................')
    
    
    
    while True:
      step = input("Enter n for north w for west e for east and s for south")
      if(step.lower()== 'n'):
          print('You will find many treasures in this path.But beware of the deadly creature')
          print("Take the treasure ")
          
          list.append('treasure')
          list.append("BonusPoints")
          print("you have got the treasure")
          print("you have got the bonus points")
          print(list)
          
      elif(step.lower()== 'w'):
          print("there is sea ")
          s = input("do u know how to swim")
          if(s.lower == 'yes ' or  s.lower == 'y'):
              print("you are safe")
              print("you have got the life points")
              list.append('lifePoints')
              
              
          else:
                print("you are dead")
                break
        
          
      elif(step.lower()== 's'):
          print("there is deadly creature")
          p = input("use points to buy the weapon .press w for weapon ")
          if(p.lower == 'w' or 'weapon'):
               if('BonusPoints' in list):
                print("you have got the weapon")
                print("you have killed the deadly creature")
                list.remove("BonusPoints")
                list.append('weapon')
                print(list)
               else:
                print("you dont have enough points to buy the weapon and deadly creature has killed you")
                break
               
          else:
                print("you are dead")
                break
            
      elif(step.lower()== 'e'):
          print('you find a new friend')
          list.remove('friendlyCreature')
          print("you are safe.your friend will take you to the exit")
          print("Bye Bye")
          print(list)
          break
      else:
          print("wrong input")
          
forest() 
print("you have got the following items")
for i in range(len(list)):
    print(list[i])
    


          
               
           
          
          
```
