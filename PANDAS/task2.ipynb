```python
import csv
import xml.etree.ElementTree as ET
import pandas as pd

root = ET.Element("products")

products = [
    {"id": "1", "name": "Laptop", "category": "Electronics", "price": "800"},
    {"id": "2", "name": "Camera", "category": "Electronics", "price": "400"},
    {"id": "3", "name": "Chair", "category": "Furniture", "price": "150"},
    {"id": "4", "name": "Bookshelf", "category": "Furniture", "price": "200"},
    {"id": "5", "name": "Headphones", "category": "Electronics", "price": "50"}
]

for product_data in products:
    product = ET.SubElement(root, "product")
    id_element = ET.SubElement(product, "id")
    id_element.text = product_data["id"]
    name_element = ET.SubElement(product, "name")
    name_element.text = product_data["name"]
    category_element = ET.SubElement(product, "category")
    category_element.text = product_data["category"]
    price_element = ET.SubElement(product, "price")
    price_element.text = product_data["price"]

tree = ET.ElementTree(root)
tree.write("products.xml")

tree = ET.parse('products.xml')
root = tree.getroot()


category_prices = {}
max_price = 0
max_price_product = None


for product in root.findall('product'):
   
    id = product.find('id').text
    name = product.find('name').text
    category = product.find('category').text
    price = float(product.find('price').text)

   
    if category in category_prices:
        category_prices[category].append(price)
    else:
        category_prices[category] = [price]

  
    if price > max_price:
        max_price = price
        max_price_product = (id, name, category, price)

average_prices = {}
for category, prices in category_prices.items():
    average_prices[category] = sum(prices) / len(prices)



with open('products.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['product_id', 'name', 'category', 'price'])
    for product in root.findall('product'):
        id = product.find('id').text
        name = product.find('name').text
        category = product.find('category').text
        price = product.find('price').text
        writer.writerow([id, name, category, price])
    
with open('category.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Category', 'Average Price'])
    for category, average_price in average_prices.items():
        writer.writerow([category, average_price])
    
with open('max_price_product.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['ID', 'Name', 'Category', ' Highest Price'])
    writer.writerow(max_price_product)
 
df = pd.read_csv('products.csv', index_col='product_id')
print(df)

print('\n')
df = pd.read_csv('category.csv',index_col='Category')
print(df)
print('\n')
df = pd.read_csv('max_price_product.csv',index_col='ID')
print(df)
```
