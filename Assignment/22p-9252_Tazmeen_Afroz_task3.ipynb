```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Read the data
df = pd.read_csv('CTI_cat_val_dataset.csv')

df = pd.DataFrame(df)
df
```

```python
#for using by default pallete you have to first import mcolors 
import matplotlib.colors as mcolors
#stor the colors values in the list
colors = list(mcolors.TABLEAU_COLORS.values())

#plot the bar graph for different values
df['category'].value_counts().plot(kind='bar', color=colors, width=0.8, figsize=(10, 5))
plt.title('Category Distribution')
plt.xlabel('Category')
plt.ylabel('Number of Occurrences')
plt.show()
```
