```python
'''1. Data Analysis Problem:
You are given a JSON file containing information about a group of people, each
represented as a dictionary with the following keys:
â€¢ "name": Name of the person (string)
â€¢ "age": Age of the person (integer)
â€¢ "country": Country of residence (string)
â€¢ "skills": List of skills the person possesses (list of strings)

{"name": "Alice", "age": 28, "country": "USA", "skills": ["Python", "JavaScript", "SQL"], "job":
"Software Engineer", "experience_years": 5},
{"name": "Bob", "age": 35, "country": "UK", "skills": ["Java", "C#", "Python"], "job": "Senior
Developer", "experience_years": 10},
{"name": "Charlie", "age": 22, "country": "Canada", "skills": ["JavaScript", "HTML", "CSS"],
"job": "Web Developer", "experience_years": 2},
{"name": "Eva", "age": 29, "country": "Australia", "skills": ["Ruby", "Swift", "Java"], "job":
"Mobile App Developer", "experience_years": 6},
{"name": "David", "age": 40, "country": "USA", "skills": ["C++", "Python", "R"], "job": "Data
Scientist", "experience_years": 15}
Write a python code to incorporate these analytical points:
1. Average Age Calculation: Calculate and print the average age of all the people in the
JSON data.
2. Skill Frequency: Determine the most common skill among all the people and print it
along with its frequency.
3. Country-wise Skills: Create a dictionary where the keys are unique countries and the
values are lists of skills possessed by people from that country.
4. Age Grouping: Group people into different age groups (e.g., 20-30, 31-40, etc.) and
count the number of people in each group. Print the age groups and the corresponding
counts.'''
```

```python
import json

data1 = {"name": "Alice", "age": 28, "country": "USA", "skills": ["Python", "JavaScript", "SQL"], "job":
"Software Engineer", "experience_years": 5}
data2 = {"name": "Bob", "age": 35, "country": "UK", "skills": ["Java", "C#", "Python"], "job": "Senior Developer", "experience_years": 10 }
data3 = {"name": "Charlie", "age": 22, "country": "Canada", "skills": ["JavaScript", "HTML", "CSS"],
"job": "Web Developer", "experience_years": 2}
data4 = {"name": "Eva", "age": 29, "country": "Australia", "skills": ["Ruby", "Swift", "Java"], "job":
"Mobile App Developer", "experience_years": 6}
data5 = {"name": "David", "age": 40, "country": "USA", "skills": ["C++", "Python", "R"], "job": "Data Scientist", "experience_years": 15}

data = [data1, data2, data3, data4, data5]
json.dump(data, open("data.json", "w"))

with open("data.json", "r") as f:
    data = json.load(f)

# 1. Average Age Calculation: Calculate and print the average age of all the people in the JSON data.
age = []
for i in range(len(data)):
    
    age.append(data[i]["age"])
avg_age = sum(age)/len(age)


print("Average age of all the people in the JSON data is: ", avg_age)

# 2. Skill Frequency: Determine the most common skill among all the people and print it along with its frequency.
skills = []
for i in range(len(data)):
    skills.append(data[i]["skills"])

skill_dict = {}

for i in range(len(skills)):
    for j in range(len(skills[i])):
        if skills[i][j] in skill_dict.keys():
            skill_dict[skills[i][j]] += 1
        else:
            skill_dict[skills[i][j]] = 1

for key, value in skill_dict.items():
    if value == max(skill_dict.values()):
        print("Most common skill among all the people is: ", key, " with frequency : ", value)

#3. Country-wise Skills: Create a dictionary where the keys are unique countries and the values are lists of skills possessed by people from that country.
country = []
for i in range(len(data)):
    country.append(data[i]["country"])

country_skills = {}

for i in range(len(country)):
    if country[i] in country_skills.keys():
        country_skills[country[i]] += skills[i]
    else:
        country_skills[country[i]] = skills[i]

print("Country wise skills possessed by people are: ", country_skills)


#4. Age Grouping: Group people into different age groups (e.g., 20-30, 31-40, etc.) andcount the number of people in each group. Print the age groups and the correspondingcounts.

age_group = {}

for i in range(len(age)):
    if age[i] >= 20 and age[i] <= 30:
        if "20-30" in age_group.keys():
            age_group["20-30"] += 1
        else:
            age_group["20-30"] = 1
    elif age[i] >= 31 and age[i] <= 40:
        if "31-40" in age_group.keys():
            age_group["31-40"] += 1
        else:
            age_group["31-40"] = 1
    elif age[i] >= 41 and age[i] <= 50:
        if "41-50" in age_group.keys():
            age_group["41-50"] += 1
        
    
     
print("Age group of people are: ", age_group)

```
