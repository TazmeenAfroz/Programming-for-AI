```python
'''2. You have multiple JSON files, each containing data about different aspects of a
company's performance. The JSON files are named sales.json, expenses.json, and
employees.json. Each file contains a list of dictionaries, with each dictionary
representing a record.
Sale.json:

Expense.json:

Employees.json:

Write a python code to incorporate these analytical points:
1. Total Sales: Calculate and print the total sales for the company across all months.
2. Total Expenses: Calculate and print the total expenses for the company across all
months.
3. Average Employee Salary: Calculate and print the average salary of employees.
4. Highest Earning Employee: Determine and print the name of the employee with the
highest salary.
5. Profit Calculation: Calculate and print the profit (sales - expenses) for each month.
6. Monthly Profit Trend: Determine and print the month with the highest profit and the
month with the lowest profit.'''
```

```python
import json
import os

employee1 = { "name":"Tazmeen" ,"dept" : "IT" , "salary" : 50000}
employee2 = { "name":"Sana" ,"dept" : "marketing" , "salary" : 70000}
employee3 = { "name":"Sara" ,"dept" : "HR" , "salary" : 90000}

employees = [employee1,employee2,employee3]

with open("employees.json","w") as f:
    json.dump(employees, f)


sale1 = { "month":"jan" ,"sales" : 50000}
sale2 = { "month":"feb" ,"sales" : 70000}
sale3 = { "month":"mar" ,"sales" : 80000}

sales = [sale1,sale2,sale3]
with open("sales.json","w") as f:
    json.dump(sales,f)

expense1 = { "month":"jan" ,"expense" : 10000}  
expense2 = { "month":"feb" ,"expense" : 60000}
expense3 = { "month":"mar" ,"expense" : 90000}

expenses = [expense1,expense2,expense3]
with open("expenses.json","w") as f:
    json.dump(expenses,f)



#1. Total Sales: Calculate and print the total sales for the company across all months
with open("sales.json","r") as f:
    sales = json.load(f)
    total_sales = 0
    for sale in sales:
        total_sales += sale["sales"]
    print("Total Sales: ",total_sales)

#2. Total Expenses: Calculate and print the total expenses for the company across all months.
with open("expenses.json","r") as f:
    expenses = json.load(f)
    total_expenses = 0
    for expense in expenses:
        total_expenses += expense["expense"]
    print("Total Expenses: ",total_expenses)

#3. Average Employee Salary: Calculate and print the average salary of employees.
with open("employees.json","r") as f:
    employees = json.load(f)
    total_salary = 0
    for employee in employees:
        total_salary += employee["salary"]
    print("Average Salary: ",total_salary/len(employees))

#4. Highest Earning Employee: Determine and print the name of the employee with the highest salary.

    max_salary = 0
    for employee in employees:
        if employee["salary"] > max_salary:
            max_salary = employee["salary"]
            max_emp = employee["name"]
    print("Highest Earning Employee: ",max_emp)
    
#5. Profit Calculation: Calculate and print the profit (sales - expenses) for each month.

dir = os.getcwd()
profit = {}
for file in os.listdir(dir):
    if file.endswith(".json"):
        file_path = os.path.join(dir, file)
        with open(file_path, "r") as f:
            data = json.load(f) 
           
            if file == "sales.json":
                for sale in data:
                    if sale["month"] not in profit:
                        profit[sale["month"]] = 0
                    profit[sale["month"]] += sale["sales"]
                    # print(profit)
            elif file == "expenses.json":
                for expense in data:
                    if expense["month"] not in profit:
                        profit[expense["month"]] = 0
                    profit[expense["month"]] -= expense["expense"]
                    # print(profit)
print("Profit: ", profit)


# #6. Monthly Profit Trend: Determine and print the month with the highest profit and the month with the lowest profit.

max_profit = 0
min_profit = 0
max_month = ""
min_month = ""
for key,value in profit.items():
    if value > max_profit:
        max_profit = value
        max_month = key
    if value < min_profit:
        min_profit = value
        min_month = key
print("Month with the highest profit: ",max_month)
print("Month with the lowest profit: ",min_month)

```
